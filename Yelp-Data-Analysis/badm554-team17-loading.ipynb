## loading dataset and inspect

```python
# review
review = pd.DataFrame()
with pd.read_json("/kaggle/input/yelp-dataset/yelp_academic_dataset_review.json", lines=True, orient='columns', chunksize=5000) as reader:   
    for chunk in reader:
        if (review.empty):
            review = chunk
        else:
            review.append(chunk)
review.describe()
```

```python
checkin = pd.read_json("/kaggle/input/yelp-dataset/yelp_academic_dataset_checkin.json", lines=True, orient='columns')
checkin.describe()
```

```python
business = pd.read_json("/kaggle/input/yelp-dataset/yelp_academic_dataset_business.json", lines=True, orient='columns')
business.describe()
```

```python
# tip
tip = pd.DataFrame()
with pd.read_json("/kaggle/input/yelp-dataset/yelp_academic_dataset_tip.json", lines=True, orient='columns', chunksize=5000) as reader:   
    for chunk in reader:
        if (tip.empty):
            tip = chunk
        else:
            tip.append(chunk)
tip.describe()
```

```python
# user
user = pd.DataFrame()
with pd.read_json("/kaggle/input/yelp-dataset/yelp_academic_dataset_user.json", lines=True, orient='columns', chunksize=5000) as reader:   
    for chunk in reader:
        if (user.empty):
            user = chunk
        else:
            user.append(chunk)
user.describe()
```
