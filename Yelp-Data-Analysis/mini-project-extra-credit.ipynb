## Q1: View state, city, postal_code to choose location based on number of restaurants

> SELECT state, city, count(business_id) as num_res, sum(stars)/count(business_id) AS avg_stars
FROM business
WHERE categories like "%Restaurant%"
GROUP BY state, city
HAVING num_res >=100 and avg_stars >= 3.7
ORDER BY avg_stars desc, num_res desc

```python
q1_1=business[business['categories'].str.contains('Restaurant',na=False, regex=True)]
q1_2 = q1_1.groupby(['state', 'city'])['business_id'].agg('count').to_frame('num_res').reset_index()
q1_3 = q1_1.groupby(['state', 'city'])['stars'].agg('sum').to_frame('sum_stars').reset_index()
q1_4 = pd.merge(left = q1_2, right = q1_3, on=['state', 'city'])
q1_4['avg_stars']= q1_4['sum_stars']/q1_4['num_res']
q1_ans = q1_4[(q1_4['num_res']>=100) & (q1_4['avg_stars']>=3.7)].sort_values(by = ['avg_stars', 'num_res'], ascending = [False, False])
```

```python
q1_ans
```

let's say we want to open a restaurant in Santa Barbara in CA!

# Q3: View attributes to decide the service attributes

> SELECT state, city, attributes,COUNT(business_id) AS num_res, sum(stars)/count(business_id) AS avg_stars
> FROM business
> WHERE categories LIKE "%Restaurants%" and categories like "%Mexican%" and city = "Santa Barbara"
> GROUP BY attributes
> HAVING avg_stars >= 3.7
> ORDER BY avg_stars desc, num_res desc
> limit 10

```python
q3_1=business[(business['categories'].str.contains('Restaurant',na=False, regex=True))& (business['city'].str.contains('Santa Barbara',na=False, regex=True))&(business['categories'].str.contains('Mexican',na=False, regex=True))]
q3_2=q3_1.copy()
q3_2['attributes'] = q3_2['attributes'].astype(str)
q3_3 = q3_2.groupby('attributes')['stars'].agg('mean').to_frame('avg_stars').reset_index()
q3_ans = q3_3[q3_3['avg_stars']>=3.7].sort_values(by=['avg_stars'], ascending = False)
```
